```python
import matplotlib.pyplot as plt
import numpy as np

# Datos con diferentes escalas
data1 = np.array([2, 5, 10, 15, 20])
data2 = np.array([50, 100, 150, 200, 250])
data3 = np.array([0.5, 1.2, 2.8, 3.5, 4.0])

# Función para normalizar datos a un rango específico
def normalize_data(data, new_min=0, new_max=1):
    min_val = np.min(data)
    max_val = np.max(data)
    normalized_data = ((data - min_val) / (max_val - min_val)) * (new_max - new_min) + new_min
    return normalized_data


def normalize_data(data, new_min=0, new_max=1):
    dim = len(data.shape)
    if dim == 1:
        min_val = np.min(data)
        max_val = np.max(data)
        normalized_data = ((data - min_val) / (max_val - min_val)) * (new_max - new_min) + new_min
    else:
        normalized_data = np.copy(data)
        columns = data.shape[1]
        for column in range(columns):
            min_val = np.min(normalized_data[:,column])
            max_val = np.max(normalized_data[:,column])
            normalized_data[:,column] = ((normalized_data[:,column] - min_val) / (max_val - min_val)) * (new_max - new_min) + new_min
    return normalized_data


# Normaliza los datos en un rango personalizado [-1, 1]
normalized_data1 = normalize_data(data1, -1, 1)
normalized_data2 = normalize_data(data2, -1, 1)
normalized_data3 = normalize_data(data3, -1, 1)

# Crea un gráfico con los datos normalizados en el nuevo rango
plt.plot(normalized_data1, label='Data 1')
plt.plot(normalized_data2, label='Data 2')
plt.plot(normalized_data3, label='Data 3')

plt.xlabel('Índice')
plt.ylabel('Valor Normalizado')
plt.title('Gráfico de Datos Normalizados en [-1, 1]')
plt.legend()
plt.show()


plt.plot(data1, label='Data 1')
plt.legend()
plt.show()

plt.plot(data2, label='Data 3')
plt.legend()
plt.show()

plt.plot(data3, label='Data 3')
plt.legend()
plt.show()
```
